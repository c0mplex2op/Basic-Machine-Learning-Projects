```python
#Importing Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
dataset = pd.read_csv('insurance.csv')
```

```python
dataset.head()
```

```python
dataset['sex'].replace({'male':0, 'female':1}, inplace=True)
dataset['smoker'].replace({'yes':1, 'no':0}, inplace=True)
dataset['region'].replace({'northeast':0, 'northwest':1, 'southeast':2, 'southwest':3}, inplace=True)
```

```python
dataset.head()
```

```python
#AGE
plt.figure(figsize=[6,6])
sns.distplot(dataset['age'])
```

```python
plt.figure(figsize=(6,6))
sns.countplot(x='sex', data=dataset)
plt.title('Sex Distribution')
plt.show()
```

```python
plt.figure(figsize=[6,6])
sns.distplot(dataset['bmi'])
```

```python
plt.figure(figsize=(6,6))
sns.countplot(x='smoker', data=dataset)
plt.title('Smoker')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.countplot(x='region', data=dataset)
plt.title('Region')
plt.show()
```

```python
X = dataset.drop(['charges'], axis=1)
Y = dataset['charges']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
X_train_prediction = model.predict(X_train)
```

```python
error_score_train = metrics.r2_score(X_train_prediction, Y_train)
error_score_train
```

```python
X_test_prediction = model.predict(X_test)
```

```python
error_score_test = metrics.r2_score(X_test_prediction, Y_test)
error_score_test
```
